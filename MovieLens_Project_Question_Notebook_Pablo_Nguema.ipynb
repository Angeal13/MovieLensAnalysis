Note that you need to use the **percent of number of releases in a year** as a parameter of popularity of a genre

Hint 1: You need to reach to a data frame where the release year is the index and the genre is the column names (one cell shows the number of release in a year in one genre) or vice versa. (Drop unnecessary column if there are any)

Hint 2: Find the total number of movies release in a year(use item dataset to get count of movies released in a particular year, store that value in a new column as 'total'). Now divide the value of each genre in that year by total to get percentage number of release in a particular year.

```python
#your code here
```

```python
items_data_new.year.value_counts()
```

```python
items_data_new.value_counts()
```

```python
big_data_new=pd.merge(items_data_new,Data_data)
movie_year=pd.DataFrame(items_data_new['movie title'].groupby(items_data_new['year']).count())
movie_year.reset_index(inplace=True)
movie_year
```

```python
items_data_new.head(5)
```

```python
new_big_data= pd.merge(items_data_new, movie_year, on='year')
new_big_data=new_big_data.drop(['movie id', 'movie title_x', 'release date', 'Sum of Genres'], axis=1)
df=new_big_data.set_index('year')
df.rename({'movie title_y': 'total'}, axis='columns', inplace=True)
df=(df.div(df['total'], axis= 0) * 100)

```

```python
genre_names=['Action', 'Adventure', 'Animation', 'Childrens', 'Comedy', 'Crime',
       'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical',
       'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']
popularity=df.groupby(['year'])[genre_names].sum()
plt.figure(figsize=(12,12), dpi=100)
plt.title("Genre Production Preference Over Time")
sns.heatmap(popularity, cmap=['yellow','orange','red'], linewidths=2)
```

```python
movie_rating=pd.merge(items_data,Data_data)
big_data=pd.merge(movie_rating,user_data)
big_data
```

```python

#your code here
most_rated=big_data.groupby('movie title').size().sort_values(ascending=False)[:25]
most_rated
plt.figure(figsize=(10,10))
most_rated.plot.barh()
plt.title('Top 25 Most Rated')
```

```python
movies_avg=big_data.groupby('movie title').agg({'rating':[np.size,np.mean]})
movies_avg_sorted=movies_avg[movies_avg['rating']['size']>=100]. sort_values([('rating','mean')],ascending=False)
top_25=movies_avg_sorted[:25]
top_25
```

```python
plt.figure(figsize=(10,10))
top_25['rating']['mean'].plot.barh()
plt.title('Top 25 AVG Rating')
```
